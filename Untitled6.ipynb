```python
class TemperatureData:
    def __init__(self, name, readings, celcilus):
        self.name = name
        self.readings = readings 
        self.celcilus = celcilus
        
    def convert_to_franheit(self):
        Faranheit = (self.celcilus * 9/5) + 32
        return Faranheit
        
    def calculate_average_temp(self):
        average_temp = sum(self.readings )/len(self.readings)
        return average_temp
    
    def find_high_temp(self):
        high_temperature = max(self.readings)
        return high_temperature
    
    def find_low_temp(self):
        low_temperature = min(self.readings)
        return low_temperature 
    
forest = TemperatureData('Amazon',[65,45,70,45,32,45,-32], 75)
print(forest.convert_to_franheit())
print(forest.find_high_temp())
print(forest.find_low_temp())
print(forest.calculate_average_temp())
```
